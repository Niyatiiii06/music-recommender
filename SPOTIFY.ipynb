```python
song_to_index= pd.Series(df.index, index= df['song_id']).to_dict()
```

```python
def recommend(song_id, n=10):
    if song_id not in song_to_index:        # check if song exists
        print("Song not found")
        return
        
    song_index = song_to_index[song_id]      # get the selected song
    selected_song = df.loc[song_index]

    print("Selected Song:")
    print(
        selected_song['track_name'],
        "-",
        selected_song['artists'],
        "| Genre:",
        selected_song['track_genre']
    )

    song_genre = selected_song['track_genre']  # Genre filtering

    filtered_df = df[
        df['track_genre'] == song_genre
    ].copy()

    filtered_df = filtered_df.reset_index(drop=True)
    X = filtered_df[features]

    scaler = StandardScaler()              # Scaling
    X_scaled = scaler.fit_transform(X)

    knn = NearestNeighbors(                     # KNN model
        n_neighbors=min(n + 1, len(filtered_df)),
        metric='cosine'
    )

    knn.fit(X_scaled)

    filtered_index = filtered_df[                      # Find selected song in filtered dataset
        filtered_df['song_id'] == song_id
    ].index[0]

    distances, indices = knn.kneighbors(                   # Find nearest neighbors
        X_scaled[filtered_index].reshape(1, -1))
    print(f"\nRecommendations for {song_id}:\n")

    for idx, dist in zip(indices[0][1:], distances[0][1:]):
        row = filtered_df.loc[idx]

        similarity = round((1 - dist) * 100, 2)

        print(
            row['track_name'],
            "-",
            row['artists'],
            "| Similarity:",
            similarity, "%"
        )
```

```python
def create_tags(row):

    tags = []

    # ---------- Energy ----------
    if row['energy'] >= 0.75:
        tags.append("high_energy")
    elif row['energy'] >= 0.45:
        tags.append("medium_energy")
    else:
        tags.append("low_energy")

    # ---------- Mood (Valence) ----------
    if row['valence'] >= 0.65:
        tags.append("happy")
    elif row['valence'] >= 0.30:
        tags.append("neutral_mood")
    else:
        tags.append("sad")

    # ---------- Danceability ----------
    if row['danceability'] >= 0.75:
        tags.append("danceable")
    elif row['danceability'] >= 0.45:
        tags.append("moderate_dance")
    else:
        tags.append("low_dance")

    # ---------- Acousticness ----------
    if row['acousticness'] >= 0.60:
        tags.append("acoustic")
    else:
        tags.append("non_acoustic")

    # ---------- Speechiness ----------
        # ---------- Speechiness (Updated) ----------
    # Thresholds: >0.66 is likely speech, 0.33-0.66 is mixed, <0.33 is musical
    if row['speechiness'] >= 0.66:
        tags.append("speech_heavy") # Rap, spoken word, podcasts
    elif row['speechiness'] >= 0.33:
        tags.append("mixed_delivery") # Pop with rap verses, rhythmic singing
    else:
        tags.append("melodic") # Pure singing (Crucial for "Infinity")   


    # ------------Tempo--------------
    if row['tempo'] < 80: tags.append("slow_tempo")
    elif row['tempo'] < 120: tags.append("mid_tempo")
    else: tags.append("fast_tempo")
        

    return " ".join(tags)
```

```python
df['tags'] = df.apply(create_tags, axis=1)
```

```python
df['text_features'] = (
    df['track_genre'].astype(str) + " " +
    df['tags'].astype(str) + " " +
    df['artists'].astype(str)
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

text_matrix = tfidf.fit_transform(
    df['text_features']
)
```

```python
# VERSION1 - AUDIO ENGINE USING MACHINE LEARNING WITH NUMERIC REPRESENTATION
def get_audio_scores(song_id, n=100):

    if song_id not in song_to_index:
        print("Song not found")
        return {}

    song_index = song_to_index[song_id]
    selected_song = df.loc[song_index]
    song_genre = selected_song['track_genre']

    filtered_df = df[
        df['track_genre'] == song_genre
    ].copy()

    filtered_df = filtered_df.reset_index()

    X = filtered_df[features]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    knn = NearestNeighbors(
        n_neighbors=min(n + 1, len(filtered_df)),
        metric='cosine'
    )

    knn.fit(X_scaled)

    filtered_index = filtered_df[
        filtered_df['song_id'] == song_id
    ].index[0]

    distances, indices = knn.kneighbors(
        X_scaled[filtered_index].reshape(1, -1)
    )

    audio_scores = {}

    for idx, dist in zip(indices[0], distances[0]):
        original_index = filtered_df.iloc[idx]['index']
        similarity = 1 - dist
        audio_scores[original_index] = similarity

    return audio_scores
```

```python
#VERSION2- NLP ENGINE WITH TEXT REPRESENTATION
def get_text_scores(song_id):

    if song_id not in song_to_index:
        print("Song not found")
        return None

    song_index = song_to_index[song_id]

    query_vector = text_matrix[song_index]

    similarities = cosine_similarity(
        query_vector,
        text_matrix
    ).flatten()

    return similarities
```

```python
# HYBRID VERSION- USE AUDIO AND TEXT VERSION TO PREDICT
def recommend_hybrid(song_id, n=10):

    audio_scores = get_audio_scores(song_id, n=100)
    text_scores = get_text_scores(song_id)

    final_scores = {}

    for idx, audio_score in audio_scores.items():

        text_score = text_scores[idx]

        final_score = (
            0.70 * audio_score +
            0.30 * text_score
        )

        final_scores[idx] = final_score

    sorted_songs = sorted(
        final_scores.items(),
        key=lambda x: x[1],
        reverse=True
    )

    results = []
    selected_index = song_to_index[song_id]

    for idx, score in sorted_songs:

        if idx == selected_index:
            continue

        row = df.loc[idx]

        results.append({
            "track_name": row['track_name'],
            "artist": row['artists'],
            "genre": row['track_genre'],
            "score": round(score * 100, 2)
        })

        if len(results) == n:
            break

    return results
```

```python
recommend_hybrid("Believer - Imagine Dragons")
```

```python
recommend_hybrid("Shape of You - Ed Sheeran")
```
